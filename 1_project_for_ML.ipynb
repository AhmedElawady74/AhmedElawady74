```python
data = pd.read_csv('/content/cars_moldova.csv')
data.info() # and here this is the time to use info() :D
```

```python
# create an interactive graph to display the number of cars by style
fig = px.bar(data, x='Style', title='Number of cars by type')
fig.update_xaxes(title='car style')
fig.update_yaxes(title='car numbers')

# display graph
fig.show()
```

```python
# this is the feature i used for the price of the car and this is the "Year"
feature_to_plot = 'Year'

# drow the scatter plot
fig = px.scatter(data, x=feature_to_plot, y='Price(euro)', title=f'Scatter Plot for {feature_to_plot} vs Price')
fig.show()
```

```python
# for bar chart:
import plotly.graph_objects as go
feature_to_plot = 'Fuel_type' # and here i used "fuel_type" wuth bar chart

# count the number of impressions for each value in the feature
value_counts = data[feature_to_plot].value_counts()

fig = go.Figure(data=[go.Bar(x=value_counts.index, y=value_counts.values)])
fig.update_layout(title=f'Bar Chart for {feature_to_plot}')
fig.show()
```

```python
# another scatter plot to display a relationship between two features (Engine Capacity(cm3) and Distance)
fig = px.scatter(data, x='Engine_capacity(cm3)', y='Distance', title='Scatter Plot for Engine Capacity vs Distance')
fig.show()
```

```python
# plotting the price distribution
plt.figure(figsize=(12, 6))
sns.histplot(data['Price(euro)'], bins=30, kde=True)
plt.title('Price Distribution')
plt.xlabel('Price (Euro)')
plt.ylabel('Number of Cars')
plt.show()

plt.figure(figsize=(12, 6)) # plotting engine capacity vs price untill the end for the next comment
sns.scatterplot(x='Engine_capacity(cm3)', y='Price(euro)', data=data, alpha=0.5)
plt.title('Engine Capacity vs Price')
plt.xlabel('Engine Capacity (cm^3)')
plt.ylabel('Price (Euro)')
plt.show()

plt.figure(figsize=(12, 6)) # plotting the distribution of manufacturing years
sns.countplot(x='Year', data=data, palette='viridis', hue='Year', legend=False)
plt.title('Distribution of Manufacturing Years')
plt.xlabel('Year of Manufacture')
plt.ylabel('Number of Cars')
plt.xticks(rotation=90)
plt.show()

# displaying some basic statistics about numeric data
numeric_data = data.select_dtypes(include=['int', 'float'])
statistics = numeric_data.describe()
print(statistics)
```
